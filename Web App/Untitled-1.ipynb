```python
import pandas as pd
import os
import datetime
```

```python
news_path = r"C:\Users\user\Desktop\done\IL"
all_files = [f for f in os.listdir(news_path) if f.endswith('.csv')]

dfs = []
for file in all_files:
    file_path = os.path.join(news_path, file)
    print(file_path)
    df = pd.read_csv(file_path)

    # df['date'] = pd.to_datetime(df['date'], format="%Y-%m-%d")
    df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d', errors='coerce')
    dfs.append(df)

news_df = pd.concat(dfs, ignore_index=True)
news_df['date'] = pd.to_datetime(news_df['date'], format='%Y-%m-%d', errors='coerce')
news_df['date'] = news_df['date'].dt.date
# new_df.drop(columns=['Unnamed: 0'], inplace=True)  # If you need to drop an extra index column
news_df
```
